# Subplots

Subplots nos sirve, al igual que subplot, para trabajar multiples graficas en un solo lienzo. Solo que ***Subplots*** aprovecha las bondades de trabajar con el metodo orientado a objetos de matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y, 'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
#Accedemos a los objetos a través de indices
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

También podemos predefinir los axes dentro de una tupla

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
#Accedemos a los objetos a través de indices
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
```

O simplemente declararlos dentro de subplots, y por defecto se comprende que el primer número corresponde a las filas y el segundo a las columnas

```python
fig, axes = plt.subplots(2, 4)
axes[0,0].plot(x, np.cos(x))
axes[0,1].plot(x, np.sin(x),'y')
axes[0,2].plot(x, np.tan(x),'r')
axes[0,3].plot(x, np.cos(x)**2,'g')
```

Volvemos a definir previamente la salida del arreglo en tuplas

```python
fig, ((ax1,ax2,ax3,ax4), (ax5,ax6,ax7,ax8)) = plt.subplots(2, 4)
ax1.plot(x, np.cos(x))
ax2.plot(x, np.sin(x),'y')
ax3.plot(x, np.tan(x),'r')
ax4.plot(x, np.cos(x)**2,'g')
#Manipulamos la apariencia de la grafica
fig.tight_layout()
```

```python
fig, ((ax1,ax2,ax3,ax4), (ax5,ax6,ax7,ax8)) = plt.subplots(2, 4)
ax1.plot(x, np.cos(x))
ax2.plot(x, np.sin(x),'y')
ax3.plot(x, np.tan(x),'r')
ax4.plot(x, np.cos(x)**2,'g')

ax5.plot(x, np.cos(y),'b')
ax6.plot(x, np.sin(y),'violet')
ax7.plot(x, np.tan(y),'black')
ax8.plot(x, np.cos(y)**2,'pink')

fig.tight_layout()
```
