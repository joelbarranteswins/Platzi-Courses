```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
```

### Comando Set

**Set** es un comando muy util que permite customizar de manera sencilla nuestros gr√°ficos. Ya sean los estilos, paleta de colores, fuente, entre otras.

[seaborn.set_theme](https://seaborn.pydata.org/generated/seaborn.set_theme.html#seaborn.set_theme)

[seaborn.color_palette](https://seaborn.pydata.org/generated/seaborn.color_palette.html#seaborn.color_palette)

```python
sns.set(style='dark', palette='dark', font="Verdana", font_scale=1 )
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
```
