## Contexto a las grficas

```python
fig, axes = plt.subplots(1,2,figsize=(5,5))
axes[0].plot(x,y_2, label="$sin(x)$")
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()
axes[1].plot(y_2,x)
axes[1].set_title('Relacion Y - X')
axes[1].set_ylabel('X')
axes[1].set_xlabel('Y')
```

```python
plt.plot(x,y_2,label='$sin(x)$')
plt.title('Este es un ttulo')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='lower left', bbox_to_anchor=(0.3,0.2))
plt.show()
```

## Grficos de Distribucin

```python
sns.histplot(data=tips, x='tip',bins=15, cumulative=False, hue='sex', stat='frequency')
# cumulative = frecuencia acumulada v:
plt.show()
```

```python
sns.histplot(data=tips, x='tip',bins=15, cumulative=False, hue='sex', stat='probability')
# `stat` must be one of ['count', 'frequency', 'density', 'probability', 'proportion', 'percent'],v:
plt.show()
```

```python
sns.histplot(data=tips, x='tip',bins=15, cumulative=False, hue='sex', stat='density')
# `stat` must be one of ['count', 'frequency', 'density', 'probability', 'proportion', 'percent'],v:
plt.show()
```

```python
sns.histplot(data=tips, x='tip',bins=15, cumulative=False, hue='sex', stat='count', multiple='dodge')
# multiple = ['layer', 'stack', 'fill', 'dodge']
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, shade=True,bw_adjust=1)
#para pintar shade :v
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex',stat='count')
plt.show()
#escalonado, esta feito kkkkk
```

```python
sns.displot(data=tips, x='tip', hue='sex', kind='hist',multiple='stack')
plt.show()
#displot es el general :v
```
