## INTRODUCCIÓN 

La economía de EE. UU. se debilitó hacia fines de 2020. Cifras oficiales mostraban la disminución del empleo y las ventas minoristas en Diciembre, lo que no proyectaba una aceleración de la economia. Aunque la tasa de nuevas infecciones comenzó a disminuir hacia fines de Diciembre, un aumento en los viajes de vacaciones combinado con nuevas cepas del virus resultaron en otro aumento de infecciones a principios de Enero.  A pesar de esta situación a finales de Enero el gasto del consumidor aumento, especialmente en los hogares de bajos ingresos, esto debido a la creciente confianza en superar la crisis, la renovación del estímulo gubernamental a finales de diciembre, y la expectativa de más estímulos por venir.                                                                       Gran parte de la perturbación de la economía estadounidense provino del debilidad gasto de los consumidores en servicios como restaurantes, líneas aéreas, hoteles y tiendas minoristas, actividades que implican cierto grado de interacción social y un riesgo de infección. Por otro lado, algunas partes de la economía tuvo un buen desempeño, como el gasto en bienes duraderos como automóviles y bienes digitales, así como la actividad en el mercado inmobiliario. En el comercio minorista, la caída de las compras en la tienda fue compensado parcialmente por un fuerte aumento en las compras en línea. Esto era consistente con otros aspectos de la vida que involucran más en interacción en linea como trabajar desde casa, entretenerse desde casa y aprender desde casa. Eventualmente, el virus será suprimido lo suficiente como para permitir un retorno a un nivel de actividad económica anterior a la pandemia. Cuándo llegue la recuperación, podemos esperar que los ingresos aumenten, los hogares dejen de ahorrar una parte tan grande de sus ingresos y, en cambio, gasten más en servicios orientados al consumidor, como restaurantes y viajes. Este cambio de comportamiento contribuirá en gran medida para impulsar la tasa de crecimiento económico. La perturbación del mercado laboral será un problema a más largo plazo: incluso una recuperación robusta a finales de este año es poco probable que alivie el desempleo entre muchos ex empleados de industrias orientadas al consumidor.  En la era post-COVID, es probable que algunos de los cambios estructurales en la economía de  EE. UU. que ocurrieron en 2020 se mantendran.  Estos podrían incluir más trabajo desde casa y una consiguiente disminución de la demanda de espacio para oficinas; más compras desde casa y la consiguiente disminución del número de  minoristas en tiendas; y más entretenimiento en el hogar y el consiguiente aumento de servicios de streaming y juegos. Despues de este análisis general y muy resumido de las condiciones de la economia americana , y de mostrar las circunstantacias que retaran la industria de retail en los proximos años, no podemos desvirtuar el atractivo mercado que esta industria representa.
```
facecolor = '#D5F5E3'
color_bars = '#73C6B6'
txt_color1 = '#2C3E50'
txt_color2 = '#A93226'

fig, ax = plt.subplots(1, figsize=(18,6))
ax.set_facecolor(facecolor)
# 'auto', 'sturges', 'fd', 'doane', 'scott', 'rice' or 'sqrt'
n, bins, patches = plt.hist(df['Sales'], color=color_bars, bins='doane')
plt.ylabel('Q Companies', c=txt_color2, fontsize=14)
plt.xlabel('Sales', c=txt_color2, fontsize=14)


#grid
minor_locator = AutoMinorLocator(2)
plt.gca().xaxis.set_minor_locator(minor_locator)
plt.grid(which='minor', color=facecolor, lw = 0.5)
xticks = [(bins[idx+1] + value)/2 for idx, value in enumerate(bins[:-1])]
xticks_labels = [ "{:.0f}-{:.0f}".format(value, bins[idx+1]) for idx, value in enumerate(bins[:-1])]
plt.xticks(xticks, labels=xticks_labels, c=txt_color1, fontsize=13)
# remove major and minor ticks from the x axis, but keep the labels
ax.tick_params(axis='x', which='both',length=0)
# remove y ticks
plt.yticks([])

# Hide the right and top spines
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
for idx, value in enumerate(n):
    if value > 0:
        plt.text(xticks[idx], value+0.3, int(value), ha='center', fontsize=16, c=txt_color1)

plt.title('Companies per Sales', loc = 'left', fontsize = 20, c=txt_color1)
plt.tight_layout()
plt.savefig('costs.png', facecolor=facecolor)
```

Otra&nbsp;oportunidad&nbsp;de&nbsp;nuevos&nbsp;negocios&nbsp;es&nbsp;la&nbsp;categoria&nbsp;Drug&nbsp;Stores,&nbsp;definiendo&nbsp;un&nbsp; modelo&nbsp;de&nbsp;negocio&nbsp;que&nbsp;aporte&nbsp;valor superior a&nbsp;los&nbsp;consumidores y enfocarlo a tendencias de proximidad, ya que permite una expansión amplia que se traduce en mayores ventas por tienda, como se observaba en el apartado de crecimiento de numero de tiendas.

Por último, si lo que se pretende es mejorar el modelo de negocio actual se recomienda canalizar esfuerzos hacia el comercio electronico. No se ve la necesidad de crear un nuevo negocio para incursionar en esta categoria, cuando se puede aprovechar la infraestructura logisticas existente y dirigir los montos de inversión a la construcción de una plataforma de calidad que permita un servicio omnicanal fuerte como factor diferenciador en el mercado.
