```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
x = np.linspace(-5,5, num=N)
```

## Funci√≥n lineal
$$y=mx+b$$

```python
def f(x):
  return x

plt.plot(x, f(x))
plt.grid()
```

## Funci√≥n escal√≥n o de Heaviside
se utiliza para categorizar

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
def H(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x>=0:
      Y[idx]=1
  return Y
    

N=1000


y = H(x)

plt.plot(x,y)
plt.grid()
```

## Funci√≥n sigmoide
se utiliza para probabilidades

$$f(x)=\frac{1}{1-e^{-x}}$$

```python
def f(x):
  return 1/(1 + np.exp(-x))
    

N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
```

## Funci√≥n tangente hiperb√≥lica

$$f(x)=\frac{2}{1+e^{-2x}}-1$$

```python
def f(x):
  return np.tanh(x)
    

N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
```

## Funci√≥n ReLU

si es negativo X, entonces el valor en Y es cero, entonces si X es positivo toma su propio valor en Y

$$R(x)=max(0,x)$$

```python
def f(x):
  return np.maximum(x,0)
    

N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
```
