```python
import pandas as pd
pd.__version__
```

```python
df = pd.read_csv('../data/covid_19_data.csv')
df.sample(10)
```

```python
#transforma la columna observation en datetime
df['ObservationDate'] = pd.to_datetime(df['ObservationDate'])
```

```python
df.dtypes
```

```python
#obtener las columnas del dataframe
list(df)
```

```python
df = df[[
    'ObservationDate',
    'Country/Region',
    'Confirmed',
    'Deaths',
    'Recovered'
    ]]
df
```

```python
# sumar segun las fechas
df_time = df.groupby('ObservationDate').sum()
df_time.head(5)
```

```python
df1 = df_time['Confirmed'].iloc[10:15]
df1
```

```python
df2 = df_time['Deaths'].iloc[12:17]
df2
```

```python
df1 - df2
```

```python
# numero de casos por dia
df_time.diff()
```

```python
# numero de casos en promedio por dia
df_time.diff().mean()
```

```python
df_diff = df_time.diff()
df_diff
```

```python
df_time.head(1).to_dict()
```

```python
df.isnull().sum()
```

```python
df_diff = df_diff.fillna({'Confirmed': 555.0,
 'Deaths': 17.0,
 'Recovered': 28.0})
```

```python
df_diff
```

```python
#otra forma de introducir datos es
df_diff.fillna(df_time.head(1))
```

```python
# volver a sumar dia a dia los numeros de casos
df_diff.cumsum()
```

```python
# suma de los numeros de valores de varios dias
df_diff.resample('7D').sum()
```

```python
# suma de los numeros de valores en los domingos
df_diff.resample('W-sun').sum()
```

```python
# suma de los numeros de valores por meses
df_diff.resample('M').sum()
```

```python
# ver el numero de valores en cada mes
df_diff.resample('M').count()
```

```python
# ver el promedio de los valores en cada mes
df_diff.resample('M').mean()
```


### Algunas opciones de resample
Alias - Description

B - Business day

D - Calendar day

W - Weekly

M - Month end

Q - Quarter end

A - Year end

BA - Business year end

AS - Year start

H - Hourly frequency

T, min - Minutely frequency

S - Secondly frequency

L, ms - Millisecond frequency

U, us - Microsecond frequency

N, ns - Nanosecond frequency
