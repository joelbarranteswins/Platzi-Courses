```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
countries = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1000, 800, 900, 1000, 300]
```

```python
plt.bar(countries, population, width=(0.5), color=['red', 'blue','orange', 'green', 'yellow'])
plt.title('Population by country')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(np.arange(5),('India','Japan','Mexico','Colombia','Germany'), rotation=45)
plt.show()
```

La misma grafica pero en horizontal

```python
plt.barh(countries, population)
plt.show()
```
