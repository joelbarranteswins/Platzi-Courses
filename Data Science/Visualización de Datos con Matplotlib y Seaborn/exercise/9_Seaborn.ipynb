# Seaborn

## ¿Por que Seaborn?



*   Esscrita por ***Michael Waskom***
*   Construida sobre Matplotilib, hereda todas las bondades que ya nos brindaba esa libreria

*   Optimizado para fucionar con Pandas y su estructura principal que es el DataFrame

## Ventajas de trabajar con Seaborn



*   Velocidad
*   Poco código

*   Altamente customizable a nivel de gráficas y visualizaciones

### Tipos de gráficas


*   Relplot (Relational): Scatterplot, lineplot
*   Distplot (Distributions): histplot, kdeplot, ecdfplot, rugplot

*   Catplot (Categorical): stripplot, swarmplot, boxplot, violinplot, pointplot, barplot
