```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Seaborn ya cuenta con ciertos datos precargados que nos permiten probar distintos gráficos con los dataset que ya trae por defecto. En este caso trabajamos con el Dataset Tips. 

[seaborn.load_dataset](https://seaborn.pydata.org/generated/seaborn.load_dataset.html)

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.displot(data=tips, x='total_bill')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip')
plt.show()
```

En el gráfico anterior podemos observar como Seaborn automáticamente sin tener que darle algúna instrucción de más, decidio que el histograma no era la mejor forma de visualizar los datos establecidos y eligió otra forma de poder visualizar de forma mas eficiente la relación entre el total de las cuentas y las propinas

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

EL parametro **hue** nos permite hacer u desglose o un agrupamiento utilizando alguna otra variable

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=True, palette='dark')
plt.show()
```

Con **kind** puedo decirle con que tipo de gráfico quiero trabajar, en este caso estamos haciendo la misma distribución que en el histograma pero por densidad
