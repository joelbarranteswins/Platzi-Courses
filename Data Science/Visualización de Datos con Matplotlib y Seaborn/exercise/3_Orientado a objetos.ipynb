# Metodo Orientado a objetos

El metodo orientado a objetos nos ofrece muchas más cualidades a diferencia de simplemente utilizar pyplot. Como por ejemplo la mayor personalización customización que puedo obtener

## Pyplot 


*  Rápido
*  Fácil

*   Una sola figura

## Object Oriented



*   Mayor Personalización
*   Más amigable y multiples diagramas

*   Más código

Un ***objeto*** define una figura, esa figura es un lienzo donde puedo meter graficas a las cuales las denominamos axes, quienes a su vez pueden tener diferentes ejes (axis)













```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,'b')
fig.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.55,0.4,0.3])

axes.plot(x,y,'b')
axes2.plot(y,x,'r')
axes2.set_facecolor('gray')
fig.show()
```
