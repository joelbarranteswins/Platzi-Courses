## multiples indices

```python
df.groupby(['cut', 'color']).mean()
```

```python
# to_frame convierte los datos en presentacion dataframe
df.groupby(['cut', 'color'])['price'].mean().to_frame()
```

```python
# aggregate aplica varias funciones a los datos y los establece como columnas
df.groupby(['cut', 'color'])['price'].aggregate(['min', 'mean', 'max'])
```

```python
def mean_kilo(x):
    return np.mean(x)/1000
```

```python
df.groupby(['cut', 'color'])['price'].aggregate(['min', 'mean', 'max', mean_kilo])
```

```python
dict_agg = {
    'carat': [min, max],
    'price':[np.mean, mean_kilo]
    }
dict_agg
```

```python
df.groupby(['cut', 'color']).aggregate(dict_agg)
```

## Crear filtros

```python
def f_filter(x):
    return mean_kilo(x['price']) > 4
```

```python
df.groupby('cut').filter(f_filter)
```

```python
df.groupby('cut').filter(f_filter)['cut'].unique()
```
