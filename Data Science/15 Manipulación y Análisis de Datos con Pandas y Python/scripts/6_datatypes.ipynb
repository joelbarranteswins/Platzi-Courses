# Manejando NaN

```python
#ver si una columna tiene NaN
df.isna().any()
```

```python
# si nos concetramos en la variable masa, veremos la cantidad de nan's que se tienen.
df['mass'].isna().value_counts()
```

```python
#borrando la columnas con NaN
df['mass']=df['mass'].dropna()
```

```python
''' si queremos borrar los nan de todas las variables
 axis=0 para filas y 1 para borrar  columnas con nan
 how = 'any' o 'all'. el primero borra la fila o columna si algunas entradas tienen nan, el segundo solo la borra si todos los valores son nan'''
 
df.dropna(axis= , how=)
```
