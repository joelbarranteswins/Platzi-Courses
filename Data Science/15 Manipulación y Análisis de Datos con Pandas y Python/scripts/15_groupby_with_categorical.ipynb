```python
import pandas as pd
import numpy as np
pd.__version__
```

```python
pd.options.display.float_format = '{:,.3f}'.format
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
#sirve para trabajar con variables categoricas
df['ones'] = 1
```

```python
df
```

```python
df_g = df.groupby(['sex','smoker'])[['ones']].sum()
df_g
```

```python
df_g.groupby(level=0).apply(

lambda x:
x / x.sum() * 100

)
```

## datos agrupados para grafica de distribucion

```python
# bins: son los rangon para cada barra de histograma
pd.cut(df['total_bill'],bins = 3).value_counts()
```

```python

pd.cut(df['total_bill'],bins = [3,18,35,60]).value_counts()
```

```python
df['bin_total'] = pd.cut(df['total_bill'],bins = [3,18,35,60])
df
```

```python
df.groupby(['time','bin_total'])[['ones']].count()
```

```python
df.groupby(['time','bin_total'])[['ones']].count().groupby(level=0).apply(

lambda x:
x / x.sum() * 100

)
```

```python
#otra manera de calcular los porcentajes
df.groupby(['time'])['bin_total'].value_counts(normalize=True).to_frame()*100
```
